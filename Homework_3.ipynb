```python
#set up plotting figures
import matplotlib.pyplot as plt
%matplotlib notebook

#import data set
import pandas as pd
data = pd.read_csv('temperatures.csv')
```

```python
#Birthday high and low temps
select = (data['month'] == 8) & (data['day'] == 14) & (data['year'] == 1996)
print("On my birthday, the lowest temperature was ", data[select]['tmin'].values[0])
print("On my birthday, the highest temperature was ", data[select]['tmax'].values[0])
```

```python
# select month, idxmin and max
# run through months
for month in range(1, 13) :
    #monthData is selected data for a particular month
    monthSelect = (data['month'] == month)
    monthData = data[monthSelect]
    monthHigh = monthData['tmax'].max()
    print("Highest recorded temperature is ", monthHigh, "for ", month)
```

```python
# Run through months
for month in range(1, 13) :
    #monthData is selected data for each month
    monthSelect = (data['month'] == month)
    monthData = data[monthSelect]
    #get number of days for month (varies)
    high = monthData['day'].max()
    low = monthData['day'].min()
    #run through each day of the month
    for day in range(low, high) :
        dayData = monthData[monthData['day'] == day]
        coldestTemp = dayData['tmin'].min()
        index = dayData['tmin'].idxmin()
        year = dayData.loc[[index]]['year'].values[0]
        print("Coldest for ", month, "/", day, ": ", coldestTemp, "set in ", year)
```

```python
'''
Find the first day that has never in any year gotten below 32
Any day where the temperature has ever gotten below freezing
is not safe to plant frost-sensitive plants (cost is not worth
the price of killing your crops were it to be an outlier year)
'''
for month in range(1, 13) :
    #monthData is selected data for each month
    monthSelect = (data['month'] == month)
    monthData = data[monthSelect]
    #get number of days for month (varies)
    high = monthData['day'].max()
    low = monthData['day'].min()
    #run through each day of the month
    isLooping = True
    for day in range(low, high) :
        dayData = monthData[monthData['day'] == day]
        coldestTemp = dayData['tmin'].min()
        if(coldestTemp > 32) :
            isLooping = False
            print("Earliest non-freezing day is", month, "/", day)
            break
    if not isLooping:
        break
```

```python
'''
Time of year when winter clothes should be ready (list assumptions)
Assumptions: winter clothes are necessary below 40 degrees, historical mean temperature per day is
sufficient to determine period of time until winter clothes are no longer necessary
'''
for month in range(1, 13) :
    #monthData is selected data for each month
    monthSelect = (data['month'] == month)
    monthData = data[monthSelect]
    #get number of days for month (varies)
    high = monthData['day'].max()
    low = monthData['day'].min()
    isLooping = True
    for day in range(low, high) :
        dayData = monthData[monthData['day'] == day]
        coldestTemp = dayData['tmin'].mean()
        if(coldestTemp > 40) :
            isLooping = False
            print("Have winter clothes until", month, "/", day)
            break
    if not isLooping:
        break
```

```python
'''
Time of year when summer clothes should be ready (list assumptions)
Assumptions: summer clothes are necessary above 75 degrees, historical mean temperature per day is
sufficient to determine period of time by which to have summer clothes
'''
for month in range(1, 13) :
    #monthData is selected data for each month
    monthSelect = (data['month'] == month)
    monthData = data[monthSelect]
    #get number of days for month (varies)
    high = monthData['day'].max()
    low = monthData['day'].min()
    isLooping = True
    for day in range(low, high) :
        dayData = monthData[monthData['day'] == day]
        hottestTemp = dayData['tmax'].mean()
        if(hottestTemp > 75) :
            isLooping = False
            print("Get summer clothes ready at", month, "/", day)
            break
    if not isLooping:
        break
```
